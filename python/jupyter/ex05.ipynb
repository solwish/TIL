# xlwings 라이브러리

+ 엑셀 프로그램에 대한 자동화 라이브러리
 - 타 엑셀 라이브러리들은 엑셀 프로그램과의 연동이 아니라, 엑셀 파일 포맷을 지원하는 형태
+ 엑셀을 듸어놓고, 파이썬을 통해 작업가능(가져오기/변경 등)
+ 엑셀의 매크로 기능 파이썬으로 구현 지원(VBA 대체 가능)
+ Anaconda Python 에 포함( pip install xlwings 또는 conda install xlwings )

```python
import xlwings as xw
```

```python
xw.__version__
```

```python
data = [
    ['apple', 'banana', 'berry'],
    [100, 200, 20],
    [12, 13, 53],
]
data
```

```python
xw.view(data)
```

```python
import numpy as np
```

```python
data = np.random.rand(3, 5)
data
```

```python
xw.view(data, xw.sheets.active)  # 현재 active 되어있는 액셀 창에서 변경됨
```

```python
import pandas as pd
```

```python
url = 'https://finance.naver.com/marketindex/exchangeList.nhn'
```

```python
df_list = pd.read_html(url)
len(df_list)
```

```python
df = df_list[0]
print(df.shape)
df.head()
```

```python
df.columns = [
    '통화명',
    '매매기준율',
    '현찰 - 사실 때',
    '현찰 - 파실 때',
    '송금 - 보내실 때',
    '송금 - 받으실 때',
    '미화환산율'
]
df.head()
```

```python
df.set_index('통화명')
```

```python
df = df.set_index('통화명')
df.head()
```

```python
xw.view(df, xw.sheets.active)
```
